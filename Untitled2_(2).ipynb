```python
#importing required library
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import seaborn as sns
```

```python
url="https://www.timeanddate.com/weather/india/lucknow/historic"
source = requests.get(url).text
# Create a BeautifulSoup object to parse the HTML content using the 'lxml' parser
soup = BeautifulSoup(source, 'lxml')

print(soup.prettify())
```

```python
#finding the table in the website
table=soup.find("table", class_="zebra tb-wt fw va-m tb-hover")
print(table.prettify())
```

```python
# Assuming you already have the 'table' object as obtained in your previous code

# Find the second header row (skip the first header row)
header_rows = table.find("thead").find_all("tr")
second_header_row = header_rows[1]

# Extract the text from each header cell in the second header row and store it in a list
column_headers = []
for header_cell in second_header_row.find_all("th"):
    column_headers.append(header_cell.text.strip())

# Now 'column_headers' will contain the column names you're looking for
df = pd.DataFrame(columns=column_headers)
# Now 'df' is a DataFrame with the column headers as columns
df
```

```python
# Find the 'tbody' (table body) element within the previously located 'table'
tbody = table.find("tbody")
data = []

if tbody:
    # Find all 'tr' (table row) elements within the 'tbody'
    rows = tbody.find_all("tr")

    # Loop through each 'tr' element to extract row data
    for row in rows:
        # Find the first 'th' element (table header) and extract its text
        header = row.find("th").text.strip()

        # Extract text from each 'td' (table data) element within the row and strip any leading/trailing whitespace
        row_data = [header] + [td.text.strip() for td in row.find_all("td")]

        # Append the extracted row data to the 'data' list
        data.append(row_data)

# Create a Pandas DataFrame using the extracted data and the previously defined 'header' (table header)
df = pd.DataFrame(data, columns=column_headers)
# Print the resulting DataFrame
df
```

```python
# Identify columns with non-empty headers and keep them
df = df.loc[:, df.columns.str.strip() != '']
# Print the DataFrame after removing empty header columns
df
```

```python
time_data = []
temp = []
wind = []
humidity = []
baro = []
visibility = []
weather = []  # to store weather conditions

for row in X:
    # Extracting time
    try:
        time_data.append(row[0])
    except ValueError:
        print(f"Error with time value: {row[0]}")
        time_data.append(None)

    # Extracting temperature value from '81°F'
    try:
        temp_val = row[1].replace('°F', '').strip()
        temp.append(int(temp_val))
    except ValueError:
        print(f"Error with temperature value: {row[1]}")
        temp.append(None)

    # Extracting weather conditions
    weather.append(row[2])

    # Extracting wind speed from '7 mph'
    try:
        wind_speed = row[3].split()[0]
        wind.append(int(wind_speed))
    except ValueError:
        print(f"Error with wind value: {row[3]}")
        wind.append(None)

    # Extracting humidity from '89%'
    try:
        humidity_val = row[4].strip('%')
        humidity.append(int(humidity_val))
    except ValueError:
        print(f"Error with humidity value: {row[4]}")
        humidity.append(None)

    # Extracting barometer value from '29.59 "Hg'
    try:
        baro_val = row[5].replace('"Hg', '').strip()
        baro.append(float(baro_val))
    except ValueError:
        print(f"Error with barometer value: {row[5]}")
        baro.append(None)

    # Extracting visibility from '2 mi'
    try:
        visibility_val = row[6].replace('mi', '').strip()
        visibility.append(int(visibility_val))
    except ValueError:
        print(f"Error with visibility value: {row[6]}")
        visibility.append(None)
```

```python
data = {'Time': pd.Series(time_data),
        'Temp': pd.Series(temp),
        'Wind': pd.Series(wind),
        'Weather': pd.Series(weather),
        'Humidity': pd.Series(humidity),
        'Barometer': pd.Series(baro),
        'visibility': pd.Series(visibility)}

cleaned_df = pd.DataFrame(data)
cleaned_df
```

```python
def calculate_median(data_list):
    sorted_list = sorted(data_list)
    n = len(sorted_list)

    if n % 2 == 1:
        return sorted_list[n // 2]
    else:
        left_mid = sorted_list[(n - 1) // 2]
        right_mid = sorted_list[n // 2]
        return (left_mid + right_mid) / 2

temp_median = calculate_median(temp)
wind_median = calculate_median(wind)
humidity_median = calculate_median(humidity)
baro_median = calculate_median(baro)
visibility_median = calculate_median(visibility)

print(f"Median Temperature: {temp_median:.2f}°F")
print(f"Median Wind Speed: {wind_median:.2f} mph")
print(f"Median Humidity: {humidity_median:.2f}%")
print(f"Median Barometer: {baro_median:.2f} \"Hg")
print(f"Median Visibility: {visibility_median:.2f} mi")
```

```python
from collections import Counter

def calculate_mode(data_list):
    counter = Counter(data_list)
    max_count = max(counter.values())
    mode_values = [key for key, value in counter.items() if value == max_count]
    return mode_values

temp_mode = calculate_mode(temp)
wind_mode = calculate_mode(wind)
humidity_mode = calculate_mode(humidity)
baro_mode = calculate_mode(baro)
visibility_mode = calculate_mode(visibility)

print(f"Mode Temperature: {', '.join(map(str, temp_mode))}°F")
print(f"Mode Wind Speed: {', '.join(map(str, wind_mode))} mph")
print(f"Mode Humidity: {', '.join(map(str, humidity_mode))}%")
print(f"Mode Barometer: {', '.join(map(str, baro_mode))} \"Hg")
print(f"Mode Visibility: {', '.join(map(str, visibility_mode))} mi")
```

```python
#Variace: In Lucknow, weather conditions exhibit varying degrees of variability across different parameters. The variance of temperature is relatively low at 10.19°F^2, suggesting that Lucknow enjoys a relatively stable climate with minimal temperature fluctuations. Wind speed, on the other hand, shows a moderate level of variability with a variance of 8.04 mph^2, indicating occasional gusts but generally calm conditions. Humidity levels exhibit significant fluctuations, as evidenced by the high variance of 38.64%^2, possibly due to seasonal changes or regional weather patterns. Barometric pressure remains relatively steady, with an exceptionally low variance of 0.00118 "Hg^2. Visibility in Lucknow experiences moderate variability, with a variance of 0.29 mi^2, implying occasional fluctuations due to factors like fog or weather events. These variance values collectively provide valuable insights into the local weather patterns, highlighting the overall stability of temperature and pressure while pointing out the noteworthy variability in humidity, wind speed, and visibility that residents and meteorologists in Lucknow should be aware of.
```

```python
def mean(data):
    return sum(data) / len(data)

temp_mean = mean(temp)
wind_mean = mean(wind)
humidity_mean = mean(humidity)
baro_mean = mean(baro)
visibility_mean = mean(visibility)

def variance(data, mean_value):
    return sum([(x - mean_value) ** 2 for x in data]) / len(data)

temp_variance = variance(temp, temp_mean)
wind_variance = variance(wind, wind_mean)
humidity_variance = variance(humidity, humidity_mean)
baro_variance = variance(baro, baro_mean)
visibility_variance = variance(visibility, visibility_mean)

print(f"Variance of Temperature: {temp_variance:.2f}°F^2")
print(f"Variance of Wind Speed: {wind_variance:.2f} mph^2")
print(f"Variance of Humidity: {humidity_variance:.2f}%^2")
print(f"Variance of Barometer: {baro_variance:.5f} \"Hg^2")
print(f"Variance of Visibility: {visibility_variance:.2f} mi^2")
```

```python
#Standard deviation - In Lucknow, the standard deviation values for various weather parameters offer additional insights into the city's climate. The standard deviation of temperature, at 3.19°F, indicates that temperature fluctuations are relatively modest, suggesting a degree of consistency in daily temperature patterns. Wind speed exhibits a standard deviation of 2.84 mph, reflecting a moderate level of variability, with occasional gusts but generally manageable wind conditions. Humidity levels demonstrate a standard deviation of 6.22%, signifying noticeable fluctuations, likely influenced by seasonal changes and local weather phenomena. The standard deviation for barometric pressure is 0.03440 "Hg, highlighting the stability of atmospheric pressure in Lucknow. Visibility, with a standard deviation of 0.54 mi, showcases moderate variability, occasionally impacted by factors such as fog or precipitation. These standard deviation values collectively provide a comprehensive picture of Lucknow's weather dynamics, emphasizing the city's relatively consistent temperatures and barometric pressure, alongside notable variability in humidity, wind speed, and visibility, all of which contribute to the city's unique climatic character.
```

```python
import math

def standard_deviation(variance_value):
    return math.sqrt(variance_value)

temp_std_dev = standard_deviation(temp_variance)
wind_std_dev = standard_deviation(wind_variance)
humidity_std_dev = standard_deviation(humidity_variance)
baro_std_dev = standard_deviation(baro_variance)
visibility_std_dev = standard_deviation(visibility_variance)

print(f"Standard Deviation of Temperature: {temp_std_dev:.2f}°F")
print(f"Standard Deviation of Wind Speed: {wind_std_dev:.2f} mph")
print(f"Standard Deviation of Humidity: {humidity_std_dev:.2f}%")
print(f"Standard Deviation of Barometer: {baro_std_dev:.5f} \"Hg")
print(f"Standard Deviation of Visibility: {visibility_std_dev:.2f} mi")
```
